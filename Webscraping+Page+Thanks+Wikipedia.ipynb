# Webscraping wikipedia gratidão


```python
from requests import get
```

```python
url = 'https://pt.wikipedia.org/wiki/Especial:Registo?type=thanks&limit=500'
#usa o método get para ler a url
response = get(url)
#printa o conteúdo da url em um txt na minha tela (meu deus do céu, eu possso fazer isso c/ qq página do mundo!)
#print(response.text)
```

```python
from bs4 import BeautifulSoup
#cria o método BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
#printa o objeto BeautifulSoup que foi criado
print(type(html_soup))
```

```python
#chama o método html_soup.find_all() para para retornar uma lista com todos os agradecimentos realizados na página 
thanks_containers = html_soup.find_all ('li', class_ = 'mw-logline-thanks')
print(type(thanks_containers))
print(len(thanks_containers))
```

```python
#teste
first_thanks = thanks_containers[0]
#print(first_thanks)
print(first_thanks.bdi.text)
#print pessoa que recebeu o agradecimento
print(first_thanks.find_all('bdi', limit =2)[1].text)
```

```python
csv = 'Source;Target'
for each_container in thanks_containers: 
    #print (each_container)
    bdis =  each_container.find_all('bdi', limit =2)
    #print(bdis)
    usuario_agradece = bdis[0].text
    usuario_recebe = bdis[1].text
    #printa o nome de cada usuário que agradece e o usuário que recebe o agradecimento
    #print(usuario_agradece,"-",usuario_recebe)
    csv = csv + '\n"'+ usuario_agradece+'";"'+ usuario_recebe +'"'
print(csv)
      
```
